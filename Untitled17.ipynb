```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

rng = np.random.default_rng(42)

def simulate_race(
    N=175,
    winner_time_sec=5*3600 + 7*60 + 22,  # 5:07:22
    p_group=0.65,                        # chance next rider is in a sametime group
    mean_peloton_gap=0.7,                # seconds between riders when not sametime
    mean_dropped_gap=18.0,               # seconds between riders once dropped
    breakup_rank=50,                     # around where the peloton starts breaking
    breakup_sharpness=8.0,               # higher = sharpr transition
):

    ranks = np.arange(1, N+1)

    # llogstiic switch from peloton---dropped as rank increases
    # w ~ 1 early, 0 late
    w = 1.0 / (1.0 + np.exp((ranks - breakup_rank) / breakup_sharpness))

    times = np.zeros(N, dtype=float)
    times[0] = winner_time_sec

    for i in range(1, N):
        # probability rider i shares same time as previous (group finish)
        # higher in peloton (w near 1), lower when dropped (w near 0)
        p_same = p_group * w[i]

        if rng.random() < p_same:
            times[i] = times[i-1]  # same finishing time group
        else:
            # draw a gap from a mixture controlled by w:
            # early: small mean gap, late: larger mean gap
            # exponential is a good simple choice for gaps
            mean_gap = w[i]*mean_peloton_gap + (1-w[i])*mean_dropped_gap
            gap = rng.exponential(mean_gap)
            # enforce at least tiny positive gap if not same-time
            gap = max(gap, 0.05)
            times[i] = times[i-1] + gap

    df = pd.DataFrame({
        "rank": ranks,
        "finish_time_sec": times,
        "gap_from_winner_sec": times - times[0]
    })
    return df

df = simulate_race()

df.head()
```

```python
plt.figure(figsize=(9,5))
plt.plot(df["rank"], df["gap_from_winner_sec"])
plt.xlabel("Rank")
plt.ylabel("Gap from winner (s)")
plt.title("Fake race: peloton shelf + later gaps")
plt.show()
```

```python
from scipy.stats import expon
from scipy.optimize import minimize

times = df["finish_time_sec"].to_numpy()
delta_t = np.diff(times)

# keeping positive gaps only
delta_t = delta_t[delta_t > 0]

# robust cap
cap = np.percentile(delta_t, 99)
delta_t_fit = delta_t[delta_t <= cap]

def neg_log_likelihood(params, data):
    pi, lam_p, lam_d = params
    if not (0 < pi < 1 and lam_p > 0 and lam_d > 0):
        return np.inf

    p1 = pi * expon.pdf(data, scale=1/lam_p)
    p2 = (1 - pi) * expon.pdf(data, scale=1/lam_d)
    return -np.sum(np.log(p1 + p2 + 1e-12))

# initial guesses
pi0 = 0.7
lam_p0 = 1 / max(np.mean(delta_t_fit[delta_t_fit <= np.percentile(delta_t_fit, 30)]), 1e-3)
lam_d0 = 1 / max(np.mean(delta_t_fit[delta_t_fit >= np.percentile(delta_t_fit, 70)]), 1e-3)

res = minimize(
    neg_log_likelihood,
    x0=[pi0, lam_p0, lam_d0],
    args=(delta_t_fit,),
    bounds=[(0.01, 0.99), (1e-4, 50), (1e-4, 50)]
)

pi_hat, lam_p_hat, lam_d_hat = res.x

print("=== Fit results ===")
print("success:", res.success)
print(f"π (peloton weight) = {pi_hat:.3f}")
print(f"mean peloton gap   = {1/lam_p_hat:.2f} s")
print(f"mean dropped gap   = {1/lam_d_hat:.2f} s")
```

```python
x = np.linspace(0, max(delta_t_fit.max(), 1), 600)

pdf_mix = (
    pi_hat * expon.pdf(x, scale=1/lam_p_hat)
    + (1 - pi_hat) * expon.pdf(x, scale=1/lam_d_hat)
)

plt.figure(figsize=(9,5))
plt.hist(delta_t_fit, bins=40, density=True, alpha=0.5, label="Data (Δt)")
plt.plot(x, pdf_mix, lw=2, label="Mixture model")
plt.plot(x, pi_hat * expon.pdf(x, scale=1/lam_p_hat), "--", label="Peloton component")
plt.plot(x, (1 - pi_hat) * expon.pdf(x, scale=1/lam_d_hat), "--", label="Dropped component")
plt.xlabel("Consecutive finishing gap Δt (s)")
plt.ylabel("Density")
plt.title("Fake race: fitted two-regime peloton model")
plt.legend()
plt.show()
```

```python
post_pel = (
    pi_hat * expon.pdf(delta_t, scale=1/lam_p_hat)
) / (
    pi_hat * expon.pdf(delta_t, scale=1/lam_p_hat)
    + (1 - pi_hat) * expon.pdf(delta_t, scale=1/lam_d_hat)
    + 1e-12
)

idx = np.where(post_pel < 0.5)[0]
if len(idx) > 0:
    print(f"Estimated breakup consecutive gap ≈ {delta_t[idx[0]]:.2f} s")
else:
    print("No breakup point found with 0.5 rule.")
```
